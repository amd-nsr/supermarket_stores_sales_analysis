* Visualizing and analysis

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

* Store sales analysis based on the available items in it.

```python
sns.lmplot(x ="Store_Sales" ,y= "Items_Available", data = df, lowess = True, aspect = 3, line_kws = {'color':"black"}).set(title="Store Sales vs Items Available")
plt.show()
```

* Store sales analysis based on the number of customers visits [daily].

```python
sns.lmplot(x ="Store_Sales" ,y= "Daily_Customer_Count", data = df, lowess = True, aspect = 3, line_kws = {'color':"red"}).set(title="Store Sales vs Daily Customer Count")
plt.show()
```

* Store sales analysis based on stores areas.

```python
sns.lmplot(x ="Store_Sales" ,y= "Store_Area", data = df, lowess = True, aspect = 3, line_kws = {'color':"green"}).set(title="Store Sales vs Store Area")
plt.show()
```

```python
df = df.sort_values(by = "Daily_Customer_Count")
sns.catplot(x = "Daily_Customer_Count",y = "Items_Available", data = df, aspect =4)
plt.xticks(rotation =90)
plt.show()
```

* The median of the stores sales totaly is ["60000"]

```python
df = df.sort_values(by="Daily_Customer_Count")
sns.violinplot(y="Store_Sales", data = df)
plt.show()
```
